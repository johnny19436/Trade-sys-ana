```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure

path='C:/Users/johnn/OneDrive/桌面/repo/Data/Indicators/RSI/Dow Jones.csv'

df = pd.read_csv(path)
plt.style.use('ggplot')
figure(figsize=(8, 6), dpi=100, facecolor='w')
plt.plot(df.index,df['RSI'])
plt.suptitle('RSI Dow Jones', fontsize=20)
plt.show()
```
